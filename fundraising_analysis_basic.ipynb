```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("/content/fundraising_data.csv")

# Display dataset
print("Fundraising Contributions Data:")
display(df)

# Total raised
total_raised = df["Amount_ETB"].sum()
goal_amount = 50000  # Updated goal
progress_percentage = (total_raised / goal_amount) * 100

print(f"Total Raised: {total_raised} ETB")
print(f"Fundraising Goal: {goal_amount} ETB")
print(f"Progress: {progress_percentage:.2f}%")

# Bar Chart: Contributions by individual
plt.figure(figsize=(12, 6))
plt.bar(df["Name"], df["Amount_ETB"])
plt.xticks(rotation=90)
plt.xlabel("Contributors")
plt.ylabel("Amount (ETB)")
plt.title("Fundraising Contributions by Individual")
plt.tight_layout()
plt.show()

# Timeline Chart: Contributions over time
df["Date"] = pd.to_datetime(df["Date"])
df_sorted = df.sort_values("Date")

plt.figure(figsize=(10, 5))
plt.plot(df_sorted["Date"], df_sorted["Amount_ETB"], marker="o")
plt.xlabel("Date")
plt.ylabel("Amount (ETB)")
plt.title("Fundraising Timeline")
plt.tight_layout()
plt.show()

# Goal vs Achieved Chart
plt.figure(figsize=(6, 6))
plt.bar(["Goal", "Raised"], [goal_amount, total_raised], color=["lightgray", "skyblue"])
plt.ylabel("Amount (ETB)")
plt.title("Fundraising Goal vs Achieved")
for i, val in enumerate([goal_amount, total_raised]):
    plt.text(i, val + 2000, f"{val} ETB", ha='center', fontsize=10)
plt.tight_layout()
plt.show()
```
